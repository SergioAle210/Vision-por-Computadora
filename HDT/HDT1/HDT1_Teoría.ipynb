### Ejercicio 2
2. Considere un píxel con valor de intensidad I_in=50 en una imagen estándar de 8 bits (0−255). Se aplican dos procesos de mejora secuenciales en el siguiente orden:

* I. Corrección Gamma con γ=0.5 (para expandir sombras).
* II. Ajuste Lineal con ganancia α=1.2 y brillo β=-10 (para contrastar).
Realice los cálculos en el dominio de flotantes normalizados [0,1] como dicta la buena práctica y convierta a entero de 8 bits solo al final.
    * a. Calcule el valor final del píxel I_out.
    * b. ¿Hubo saturación (clipping) en el proceso?
    * c. Si hubiéramos realizado las operaciones usando uint8 directamente sin convertir a float (truncando decimales en cada paso intermedio), ¿cuál habría sido el error numérico resultante?


#### Normalización
El primer paso para poder realizar las transformación del valor de `I_in` al dominio de los flotantes. Lo cuál se puede hacer con la fórmula: $I_{float} = \frac{I_{in}}{255}$

*Dando como resultado:*

$I_{float}$  = 50 / 255 = 0.19607843137

$I_{float}$ = 50 / 255 = 0.196

PERO. Importante tener en cuenta que el valor de β también tiene que ser normalizado

$\beta_{float}$  = -10 / 255 = -0.03921568627

$\beta_{float}$ = -10 / 255 = -0.039

#### I.

Luego de hacer la normalización para el dominio de los flotantes se hace la corrección gamma (γ), esto se logra con la fórmula: $I_{gamma} = (I_{float})^\gamma$

$I_{gamma} = (0.196)^{0.5}$ = 0.44271887242

$I_{gamma} = (0.196)^{0.5}$ = 0.443

#### II.

Una vez normalizados los datos y habiendo hecho la corrección gama, se puede realizar el ajuste lineal con las datos calculados: $I_{lineal} = \alpha \times I_{gamma} + \beta_{float}$

$I_{lineal}$ = 1.2 * 0.443 + (-0.039) = 0.4926

$I_{lineal}$ = 1.2 * 0.443 + (-0.039) = 0.493


#### a.
Para calcular el valor final es: $I_{out} = I_{lineal} \times 255$

$I_{out}$ = 0.493 * 255 = 125.715

$I_{out}$ = 0.493 * 255 = 126

#### b.
No hubo saturación/clipping en este caso, ya que al realizar la resta en el ajuste lineal el valor se mantuvo dentro del rango de [0, 1]

#### c.
De haber hecho el cálculo sin normalizaciones el resultado sería:


$I_{gamma} = (50)^{0.5}$ = 7.07106781187

$I_{gamma} = (50)^{0.5}$ = 7


$I_{lineal}$ = 1.2 * 7 - 10 = -1.6

En este caso si hay clipping, redondeandolo para que se en el rango de [0, 255] eel resultado sería 0. Algo completamente diferente al valor obtenido ralizando las normalizaciones de valores flotantes.
