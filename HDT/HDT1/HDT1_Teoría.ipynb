## Ejercicio 1

1. Como director de un proyecto de conducción autónoma, debe dimensionar el hardware para un nuevo vehículo. El sistema utiliza 8 cámaras que capturan video a resolución 4K UHD (3840 x 2160). Debido a la necesidad de alto rango dinámico (HDR), los sensores operan a 12 bits por píxel (Raw Bayer Pattern) a 60 FPS. Métrica A: Enfocada puramente en el flujo vehicular. 

**a. Calcule el tamaño exacto de una sola imagen (frame) cruda en Megabytes (MB).**  
$\text{Pixels} = 3840\times 2160 = 8{,}294{,}400\ \text{pixeles por frame}$  

$\text{Bits/frame} = 8{,}294{,}400\times 12 = 99{,}532{,}800\ \text{bits por frame}$ 

$\text{Bytes/frame} = 99{,}532{,}800/8 = 12{,}441{,}600\ \text{bytes por frame}$

$1\ \text{MiB} = 1{,}048{,}576\ \text{bytes}$.

$12{,}441{,}600\ /\ 1{,}048{,}576 = 11.865234375\ \text{MiB}$

**Respuesta:** Un frame crudo pesa $11.865234375\ \text{MiB}$.

**b. Calcule el ancho de banda necesario (en Gbps) para transmitir el flujo de las 8 cámaras al procesador central sin compresión.**  
$\text{bps/cámara} = 99{,}532{,}800\times 60 = 5{,}971{,}968{,}000\ \text{bits/s por cámara}$

$\text{bps (8 cámaras)} = 5{,}971{,}968{,}000\times 8 = 47{,}775{,}744{,}000\ \text{bits/s}$  

$1\ \text{Gib/s} = 1{,}073{,}741{,}824\ \text{bits/s}$.

$47{,}775{,}744{,}000\ /\ 1{,}073{,}741{,}824 = 44.49462890625\ \text{Gib/s}$

**Respuesta:** Se requieren $44.49462890625\ \text{Gib/s}$ para enviar el video crudo de las 8 cámaras (sin compresión).

**c. Si su procesador tiene una memoria RAM reservada de 16 GB exclusivamente para el buffer de video, ¿cuántos segundos de historia puede almacenar antes de empezar a sobrescribir datos?**  
Primero paso el flujo total a bytes/s:

$\text{Bytes/s (8 cámaras)} = 47{,}775{,}744{,}000/8 = 5{,}971{,}968{,}000\ \text{bytes/s}$  

$16\ \text{GiB} = 17{,}179{,}869{,}184\ \text{bytes}$.

$\text{Tiempo} = 17{,}179{,}869{,}184\ /\ 5{,}971{,}968{,}000 = 2.87675171468\ \text{s}$  

**Respuesta:** puedo almacenar aproximadamente $2.88\ \text{s}$ de historia antes de sobrescribir.

**d. Basado en su resultado, ¿es viable enviar estos datos "crudos" a la nube en tiempo real usando 5G? Justifique**  
**Respuesta:** no es viable enviar el video crudo en tiempo real, porque el flujo requerido es de aproximadamente $44.49\ \text{Gib/s}$ sostenidos, muy por encima de lo que típicamente se logra como subida en 5G y, además, el enlace es variable (cobertura, movilidad, congestión).

2. Considere un píxel con valor de intensidad I_in=50 en una imagen estándar de 8 bits (0−255). Se aplican dos procesos de mejora secuenciales en el siguiente orden:

* I. Corrección Gamma con γ=0.5 (para expandir sombras).
* II. Ajuste Lineal con ganancia α=1.2 y brillo β=-10 (para contrastar).
Realice los cálculos en el dominio de flotantes normalizados [0,1] como dicta la buena práctica y convierta a entero de 8 bits solo al final.
    * a. Calcule el valor final del píxel I_out.
    * b. ¿Hubo saturación (clipping) en el proceso?
    * c. Si hubiéramos realizado las operaciones usando uint8 directamente sin convertir a float (truncando decimales en cada paso intermedio), ¿cuál habría sido el error numérico resultante?


#### Normalización
El primer paso para poder realizar las transformación del valor de `I_in` al dominio de los flotantes. Lo cuál se puede hacer con la fórmula: $I_{float} = \frac{I_{in}}{255}$

*Dando como resultado:*

$I_{float}$  = 50 / 255 = 0.19607843137

$I_{float}$ = 50 / 255 = 0.196

PERO. Importante tener en cuenta que el valor de β también tiene que ser normalizado

$\beta_{float}$  = -10 / 255 = -0.03921568627

$\beta_{float}$ = -10 / 255 = -0.039

#### I.

Luego de hacer la normalización para el dominio de los flotantes se hace la corrección gamma (γ), esto se logra con la fórmula: $I_{gamma} = (I_{float})^\gamma$

$I_{gamma} = (0.196)^{0.5}$ = 0.44271887242

$I_{gamma} = (0.196)^{0.5}$ = 0.443

#### II.

Una vez normalizados los datos and habiendo hecho la corrección gama, se puede realizar el ajuste lineal con las datos calculados: $I_{lineal} = \alpha \times I_{gamma} + \beta_{float}$

$I_{lineal}$ = 1.2 * 0.443 + (-0.039) = 0.4926

$I_{lineal}$ = 1.2 * 0.443 + (-0.039) = 0.493


#### a.
Para calcular el valor final es: $I_{out} = I_{lineal} \times 255$

$I_{out}$ = 0.493 * 255 = 125.715

$I_{out}$ = 0.493 * 255 = 126

#### b.
No hubo saturación/clipping en este caso, ya que al realizar la resta en el ajuste lineal el valor se mantuvo dentro del rango de [0, 1]

#### c.
De haber hecho el cálculo sin normalizaciones el resultado sería:


$I_{gamma} = (50)^{0.5}$ = 7.07106781187

$I_{gamma} = (50)^{0.5}$ = 7


$I_{lineal}$ = 1.2 * 7 - 10 = -1.6

En este caso si hay clipping, redondeandolo para que se en el rango de [0, 255] eel resultado sería 0. Algo completamente diferente al valor obtenido ralizando las normalizaciones de valores flotantes. 

## Ejercicio 3


Usted está programando un robot clasificador de pelotas. Tiene dos objetos:
**Escenario:**
- Pelota roja bajo el sol: $R_{rgb} = (255, 0, 0)$
- Pelota roja en la sombra: $S_{rgb} = (50, 0, 0)$



#### Distancia en espacio RGB:

La distancia euclidiana entre $R$ y $S$ en RGB es:

$d(R,S) = \sqrt {(R_1-S_1)^2 + (R_2-S_2)^2 + (R_3-S_3)^2}$

por tanto , 

$d(R,S) = \sqrt {(255-50)^2 + (0-0)^2 + (0-0)^2}$

$= \sqrt {205^2} = 205$


#### Conversión a espacio HSV (normalizado):**

Normalización: 

$$ r = \frac{R}{255}$$
$$ g = \frac{G}{255}$$
$$ b = \frac{B}{255}$$

donde , 
para $R$
$r = 1$, 
$g = 0$
$b = 0$

y $S$

$r = \frac{50}{255} = 0.196$
$g = 0$
$b = 0$


Calculamos $C_{max}$ , $C_{min}$ y $\Delta$

$C_{max} = \max(r,g,b)=1$
$C_{min} = \min(r,g,b)=0$
$\Delta = C_{max} - C_{min}=1$

entonces $V = C_{max} = 1$

$$S = \begin{cases} 0 & si C_{max}=0 \\ \frac{\Delta}{C_{max}} & si C_{max}\neq 0 \end{cases}$$



**c. Diferencia absoluta canal por canal en HSV:**

De los cálculos anteriores:
$$R_{HSV} = (0, 1, 1)$$
$$S_{HSV} = (0, 1, 50/255) \approx (0, 1, 0.196)$$

Diferencias:
$$\Delta H = |0 - 0| = 0$$
$$\Delta S = |1 - 1| = 0$$
$$\Delta V = |1 - 0.196| = 0.804$$


**d. Argumento matemático para el agrupamiento:**

RGB mezcla la cromaticidad con la iluminación, mientras que HSV separa la cromaticidad de la iluminación.

Si se observa una superficie bajo iluminación constante, el color de la superficie es constante, por lo que el valor de la cromaticidad es constante.

$$I = \alpha\rho$$
En donde, 

- $I$ es la imagen
- $\rho$ es el color aparente del objeto
- $\alpha > 0$ es un factor de sombreado/iluminación

Tomemos a la pelota 

$$(50,0,0) = \alpha(255,0,0), \quad \alpha = \frac{50}{255}$$
Es decir, mismo color con intensidad rebajada.

Por lo tanto, 

Un algoritmo de agrupamiento como k-means usa la distancia euclidiana para agrupar los puntos.

$$d_{RGB}(x,y) = ||x-y||_2$$

$$d_{RGB}((255,0,0),(50,0,0)) = \sqrt{(205)^2} = \sqrt{42025} = 205$$

El algoritmo separa por brillo, no por magnitud, por lo que se estan generando puntos sobre la recta que cruza el origen, sin embargo el cálculo de la distancia es absoluta, no alineación.


En contraste, en HSV se separa color (Hue) de intensidad (Value). Para sombras, HSV se usa precisamente porque separa brillo de color y permite tratar sombras sin "cambiar el color".

$$R_{HSV}=(0,1,1)$$
$$S_{HSV}=(0,1,50/255)$$

$$\Delta H = 0$$
$$\Delta S = 0$$

$V$ es el único valor que varía, por lo que la distancia cromática es: 

$$d_{HSV\text{ (croma)}} = \sqrt{(H_1-H_2)^2+(S_1-S_2)^2}=0$$

Por lo tanto, 

Al separar Hue y Saturation, el algoritmo agrupa y reconoce la estabilidad entre ambas.
